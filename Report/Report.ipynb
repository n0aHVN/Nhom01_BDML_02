# Phần 1: Giới thiệu


Tên đề tài: **Xây dựng mô hình dự đoán để dự đoán sự trì hoãn của các chuyến bay**

## **1.1. Giới thiệu đề tài:**
- Sự trì hoãn của chuyến bay có thể gây ra nhiều rắc rối cho các hãng hàng không, hành khách và các bên khai thác sân bay. Các hãng hàng không có thể phải chịu thêm chi phí cho nhân viên, nhiên liệu và cơ sở hạ tầng. Hành khách có thể bị gián đoạn kế hoạch đi lại và phải chịu những bất tiện khác. Các bên khai thác sân bay có thể phải đối mặt với việc tắc nghẽn, ùn ứ hàng hoá và giảm hiệu quả hoạt động. Đi kèm với đó, sự gia tăng của lượng khách du lịch trong nước và ngoài nước và sự phức tạp của hoạt động hàng không đã khiến cho tầm quan trọng của việc dự đoán sự trì hoãn của các chuyến bay ngày càng được chú ý.

- Việc giải quyết bài toán dự đoán sự trì hoãn của chuyến bay sẽ giúp các hãng hàng không, hành khách và các nhà khai thác sân bay giảm thiểu những tác động tiêu cực của sự trì hoãn, cụ thể như:
1.   Các hãng hàng không có thể sử dụng mô hình dự đoán để xác định các chuyến bay có nguy cơ bị trì hoãn cao. Điều này sẽ giúp các hãng hàng không có kế hoạch dự phòng để giảm thiểu thiệt hại do sự trì hoãn gây ra. Ví dụ, các hãng hàng không có thể điều động thêm nhân viên, nhiên liệu hoặc máy bay để giải quyết các chuyến bay bị trì hoãn.
2.   Hành khách có thể sử dụng mô hình dự đoán để biết trước khả năng chuyến bay của họ bị trì hoãn. Điều này sẽ giúp hành khách chủ động sắp xếp lịch trình của mình để tránh những bất tiện do sự trì hoãn gây ra. Ví dụ, hành khách có thể đặt phòng khách sạn hoặc vé tàu hỏa dự phòng để phòng trường hợp chuyến bay của họ bị trì hoãn.
3.   Các bên khai thác sân bay có thể sử dụng mô hình dự đoán để xác định các thời điểm và địa điểm có nguy cơ xảy ra sự trì hoãn cao. Điều này sẽ giúp họ có kế hoạch điều phối và quản lý các chuyến bay để giảm thiểu tắc nghẽn và cải thiện hiệu quả hoạt động.

- Từ đó, mục tiêu của dự án là thực hiện phân tích dữ liệu lịch sử chuyến bay để thu được những hiểu biết có giá trị và xây dựng mô hình dự đoán để dự đoán liệu chuyến bay sẽ bị trì hoãn hay không khi được cung cấp một bộ đặc điểm chuyến bay.

**Các câu hỏi cần được giải đáp:**
1.   Dự đoán chuyến bay nào sẽ bị hoãn hoặc bị huỷ
2.   Dự đoán thời gian hoãn của chuyến bay
3.   So sánh sự trì hoãn của các hãng hàng không
4.   Xác định sân bay nào bị chậm trễ nhất?
5.   Xác định những tuyến đường nào thường bị chậm trễ nhất?
6.   Xác định thời gian chậm trễ tại sân bay tại điểm khởi hành mỗi tháng
7.   Xác định thời gian trì hoãn tại sân bay tại điểm khởi hành mỗi ngày/giờ
8.   Kết luận nguyên nhân chính dẫn đến sự chậm trễ chuyến bay là gì?

Bước 2: Chỉ sử dụng những chuyến bay không bị huỷ - Cancelled -  hoặc không bị chuyển hướng - Diverted:

```python
df = df.filter((df.Cancelled == 0)&(df.Diverted == 0))
df.filter((df.Cancelled != 0)|(df.Diverted != 0)).toPandas().to_csv('../data/Cancelled_Or_Diverted_Flights.csv')
df = df.withColumn('Delayed', (df.ArrDelay >=15).cast('int'))
```

Bước 3: Khai báo dataframe thành table trong SQLContext:

```python
sqlContext.registerDataFrameAsTable(df, "df")
```

Bước 4: Tạo thêm cột TimeSlot dựa trên CRSDepTime:


-   CRSDepTime <= 800 thì TimeSlot = 1
-   800 < CRSDepTime <= 1200 thì TimeSlot = 2
-   1200 < CRSDepTime <= 1600 thì TimeSlot = 3
-   1600 < CRSDepTime <= 2100 thì TimeSlot = 4



```python
df = sqlContext.sql("select *, case \
               when CRSDepTime <= 800 then 1 \
               when 800 < CRSDepTime and CRSDepTime <= 1200 then 2 \
               when 1200 < CRSDepTime and CRSDepTime <= 1600 then 3 \
               when 1600 < CRSDepTime and CRSDepTime <= 2100 then 4 \
               else 1 end as TimeSlot \
               from df")
```

Ép kiểu dữ liệu:

```python
df = df.withColumn("FlightDate", to_timestamp(df["FlightDate"], "MM/dd/yyyy"))
```

## 6.1. Các thuật toán phân loại – Classification:
### 6.1.1. Hồi quy logistic
Hồi quy logistic là một kỹ thuật phân tích dữ liệu sử dụng toán học để tìm ra mối quan hệ giữa hai yếu tố dữ liệu. Mô hình hồi quy logistic được ứng dụng trong dự đoán đối tượng thuộc vào một trong hai nhóm. Hồi quy logistic sử dụng nguyên lí của hàm sigmoid. [2]

Hàm sigmoid là một hàm phi tuyến biến đổi đầu vào thành xác suất tương ứng với một trong hai lớp nhị phân. Hàm sigmoid được diễn giải sau đây:

$$
S(z)= \frac {1}{1+e^{-z}}
$$

Với z bất kì, hàm sigmoid trả về đầu ra là một giá trị xác suất nằm trong khoảng [0;1]. Tương tự, ta có z=Xω nếu đầu vào là một ma trận dữ liệu X và trọng số ω. [2]

Hàm mất mát của bài toán hồi quy logistic là hàm mất mát Cross-Entropy. Hàm mất mát này được định nghĩa như sau: [2]

$$
L(ω)= -1/n ∑_(i=1)^ny_i log p_i+(1-y_i )  log⁡(1-p_i )
$$

Với:
*	n: Số lượng mẫu từ tập dữ liệu dùng để huấn luyện mô hình dự đoán.
*	y_i : Giá trị thực tế từ mô hình của đầu ra thứ i.
*	p_i: Với đầu vào thứ i, p_i là xác suất dự đoán đầu vào này thuộc lớp 1.


Ưu điểm:
*	Hồi quy logistic là phương pháp dễ dàng sử dụng: Để tạo ra được mô hình máy học thì cần huấn luyện và thử nghiệm mô hình, và mô hình hồi quy logistic dễ dàng trong việc đào tạo và triển khai.
*	Hoạt động tốt với các dữ liệu có phân tách tuyến tính: Tập dữ liệu phân tách tuyến tính được hiểu như là có một đường thẳng ở giữa phân tách tập dữ liệu làm hai phần riêng biệt.

Nhược điểm:
*	Hồi quy logistic giả định tuyến tính giữa biến dự đoán và biến dự báo: Trong hiện thực, rất ít có trường hợp mà chúng ta có thể thu thập được dữ liệu phân tách tuyến tính. Do vậy, đây vừa là điểm mạnh nhưng vừa là điểm yếu của mô hình hồi quy logistic.
*	Tính chính xác thấp nếu như kích thước để huấn luyện mô hình quá nhỏ: Điều này có thể dẫn tới trường hợp mô hình bị quá khớp. Do đó, để dự đoán các kết quả mới sẽ dường như là bất khả thi với trường hợp này.

Trong đề tài này nhóm quyết định sử dụng thuật toán Hồi quy logistic để dự đoán sự chậm trễ của chuyến bay vì có thể dễ dàng triển khai và vì đầu ra của thuật toán là biến phân loại.



### 6.1.2. Cây quyết định - Decision Tree:
Cây quyết định là phương pháp phân lớp các đối tượng, là một cây phân cấp có cấu trúc và có các luật khác nhau. Các kiểu dữ liệu được sử dụng trong cây quyết định có thể đa dạng như biến số lượng, biến định danh, biến thứ tự, biến nhị phân. [3]

Thuật toán ID3 của cây quyết định là phương pháp tìm kiếm một cách tham lam từ trên xuống dưới và có thể không có quay lui. Thuật toán ID3 sử dụng Entropy và Information Gain để xây dựng nên một cây quyết định. [3]

Đối với phân phối xác suất của biến rời rạc x có thể nhận n giá trị khác nhau, x1, x2, ..., xn. Từ đó, ta sẽ có giá trị xác suất khác nhau để x có thể nhận x1, x2, ..., xn, gọi là p1, p2, ..., pn. Công thức tính p là pi = p(x = xi). Từ đó, Entropy của phân phối này được tính theo công thức sau: [3]

$$
H(p)= ∑_(n=1)^np_i  log⁡(p_i)
$$


Entropy sẽ đạt tối đa khi xác suất xảy ra của hai lớp ngang bằng với nhau. Từ đó phân loại p thành hai loại, tinh khiết và vẩn đục: [3]
*	P tinh khiết: pi = 0 hoặc pi = 1
*	P vẩn đục: pi = 0.5, khi đó hàm Entropy đạt đỉnh cao nhất

Information gain dựa vào sự giảm của hàm Entropy khi bắt đầu sự phân chia của một thuộc tính. Để xây dựng quyết định thì luôn luôn phải sử dụng các thuộc tính có information gain cao nhất. [3]

Information gain được tính toán tại mỗi nút trong cây quyết định như sau:

*	Bước 1: Gọi S là biến mục tiêu có tổng số phần tử là N, NC là số phần tử thuộc lớp C cho trước.
$$
H(S)= -∑_(C=1)^C(N_C/N)  log⁡(N_C/N)
$$
*	Bước 2: Tính Entropy, tại mỗi thuộc tính x, dữ liệu trong S sẽ được phân chia thành các k lớp các nốt con gọi là S1, S2, …, Sk với lượng dữ liệu trong mỗi nốt con là m1, m2, …, mk. Ta có:
$$
H(x,S)= ∑_(k=1)^km_k/n*H(S_k)
$$
*	Bước 3: Chỉ số Gain Information được tính theo công thức:
$$
G(x,S) = H(S) – H(x,S)
$$

Ưu điểm:
*	Mô hình cây quyết định là một mô hình có thể diễn giải được, dễ dàng trong việc tìm hiểu và áp dụng.
*	Có thể làm việc với nhiều loại biến khác nhau.
*	Có khả năng làm việc với dữ liệu lớn.

Nhược điểm:
*	Mô hình này phần lớn lệ thuộc vào dữ liệu huấn luyện, một sự thay đổi nhỏ sẽ khiến cả cây quyết định thay đổi.[3]

Mô hình cây quyết định được nhóm sử dụng trong đề tài vì đầu vào của dữ liệu là nhiều loại biến khác nhau, và do mô hình này có khả năng làm việc với dữ liệu lớn.

### 6.1.3. Naive Bayes
Naive Bayes là một kỹ thuật phân loại dựa trên định lý Bayes, mô hình này là một trong những mô hình tiêu biểu sử dụng hướng xác suất thống kê để phân loại, mô hình này giả định mọi biến dự đoán là độc lập với nhau. Phân loại Naïve Bayes là một mô hình học có giám sát và tính xác suất của một giả thuyết dựa trên dữ liệu cho trước [4].

Công thức của định lý Bayes:
$$
P(A│B)=(P(B│A)*P(A))/(P(B))
$$

Trong đó:
*	P(A│B): Xác suất sau, là xác suất của mục tiêu A với đặc trưng B.
*	P(B|A): Xác suất likelihood, xác suất của đặc trưng B khi biết mục tiêu A.
*	P(A): Xác suất trước của mục tiêu A.
*	P(B): Xác suất trước của đặc trưng B.


Với công thức này, áp dụng với X là các vector điều kiện đặc trưng,
$$
P(y│X)=(P(X│y)*P(y))/(P(X))
$$

thì Naïve Bayes sẽ trở thành:
$$
P(y│x_1,x_2,…,x_n )=(P(x_1│y)P(x_2│y)…P(x_n |y)P(Y))/(P(x_1 )P(x_2 )…P(x_n))
$$

Một số kiểu mô hình Naïve Bayes: [4]

*	Naïve Bayes đa thức: Mô hình đa phần được sử dụng trong phân loại văn bản. Đặc trưng đầu vào ở đây chính là tần suất xuất hiện của từ trong văn bản đó.
*	Bernoulli Naive Bayes: Bằng cách sử dụng phân phối Bernoulli, áp dụng khi các đặc trưng đầu vào chỉ nhận giá trị nhị phân 0 hoặc 1.
*	Gaussian Naive Bayes: Khi đặc trưng có giá trị liên tục, với giả định rằng các biến đều có phân phối Gaussian, ta sẽ tính toán được likelihood theo công thức:

$$
P(x_i│y)=1/√(2πσ_y^2 )  exp⁡(-(x_i-μ_y )^2/(2σ_y^2 ))
$$

Nhóm quyết định sử dụng mô hình Naïve Bayes trong đề tài này vì để dự đoán độ trễ của các chuyến bay cần mô hình có đầu ra là biến phân loại, và mô hình Naïve Bayes đáp ứng được nhu cầu.

## 6.3. Các phương pháp đánh giá và độ đo mô hình
### 6.3.1. Kiểm chứng chéo – Cross Validation:

Kiểm chứng chéo là một phương pháp kiểm tra độ chính xác của một phương pháp máy học dựa trên một tập dư liệu học cho trước. Thay vì chỉ dùng một phần dữ liệu làm tập dữ liệu học thì kiểm chứng chéo dùng toàn bộ dữ liệu để dạy cho máy. [24]

Ba loại phương pháp kiểm chứng chéo phổ biến được sử dụng rộng rãi:
*	Hold-out: phương pháp đơn giản nhất. Dữ liệu được chia một cách ngẫu nhiên thành một tập dữ liệu học và một tập dữ liệu kiểm tra. Dùng tập đầu tiên để dạy máy rồi dùng ngay tập còn lại để kiểm tra.
*	K-fold: đây là phương pháp nâng cấp của hold-out. Toàn bộ dữ liệu được chia thành K tập con. Quá trình học của máy có K lần. Trong mỗi lần, một tập con được dùng để kiểm tra và K-1 tập còn lại dùng để dạy.
*	Leave-one-out: Phương pháp này tương tự như K-Fold nhưng tối đa hóa số tập con (K = số dữ liệu).

```python
input_cols = ["Month","DayOfWeek","Airline","Origin","Dest","TimeSlot"]
output_col = ["Delayed"]

df_temp1 = sqlContext.read.format(
    "csv").option("header", "true").load("../data/Combined_Flights_2018.csv",
    inferSchema=True)
df_temp2 = sqlContext.read.format(
    "csv").option("header", "true").load("../data/Combined_Flights_2020.csv",
    inferSchema=True)
df_model = df_temp1.union(df_temp2)
```

```python
df_model = df_model.fillna(0)
df_model = df_model.filter((df_model.Cancelled == 0)&(df_model.Diverted == 0))
df_model = df_model.withColumn('Delayed', (df_model.ArrDelay >=15).cast('int'))
sqlContext.registerDataFrameAsTable(df_model, "df_model")
df_model = sqlContext.sql("select *, case \
               when CRSDepTime <= 800 then 1 \
               when 800 < CRSDepTime and CRSDepTime <= 1200 then 2 \
               when 1200 < CRSDepTime and CRSDepTime <= 1600 then 3 \
               when 1600 < CRSDepTime and CRSDepTime <= 2100 then 4 \
               else 1 end as TimeSlot \
               from df_model")
```

```python
stringIndexer = StringIndexer(inputCols=["Airline","Origin","Dest"], outputCols=["Airline_Index","Origin_Index","Dest_Index"])
df_model = stringIndexer.fit(df_model).transform(df_model)
input_cols = ["Month","DayOfWeek","Airline_Index","Origin_Index","Dest_Index","TimeSlot"]
df_model = df_model.select(input_cols+output_col+["Year"])
```

```python
assembler = VectorAssembler(inputCols=input_cols, outputCol="features")
df_model = assembler.transform(df_model)
```

```python
def removeSomeRow_df(train_df,num_of_row):
    window_spec = Window.orderBy(functions.rand())
    df_temp1 = train_df.filter(train_df.Delayed == 1)
    df_temp0 = train_df.filter(train_df.Delayed == 0).withColumn("random_value", functions.row_number().over(window_spec))
    df_temp0 = df_temp0.filter(df_temp0["random_value"]>num_of_row)
    df_temp0 = df_temp0.drop("random_value")
    return df_temp1.union(df_temp0)
```

```python
train_df = removeSomeRow_df(train_df,2000000)
print("Số chuyến bay Delay trong tập train: ",train_df.filter(train_df.Delayed == 1).count())
print("Số chuyến bay không bị Delay trong tập train: ",train_df.filter(train_df.Delayed == 0).count())
```

Ngoài ra, nhóm có thể cải thiện thêm tập dữ liệu bằng cách bổ sung thêm các chuyến bay từ các năm tiếp theo như 2019, 2020, 2021, ...
